```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
iris = sns.load_dataset("iris")
iris_=iris.copy()
species = iris.pop("species")
sns.clustermap(iris)
```

```python
set(iris_.species)
```

```python
color = []
for item in iris_.species.values:
    if item == 'setosa': color.append('grey')
    if item == 'versicolor': color.append('blue')
    if item == 'virginica': color.append('yellow')
color
        
```

```python
plt.figure(figsize=(3,3))
sns.set(color_codes=True)
iris = sns.load_dataset("iris")
iris_=iris.copy()
species = iris.pop("species")
sns.clustermap(iris, row_colors=color)
plt.show()
```
